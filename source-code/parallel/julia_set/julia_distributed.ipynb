# Requirements

```julia
using Distributed
using Plots
```

Add worker processes, the number depends on the hardware, choose it equal to the number of physical cores.

```julia
addprocs(4)
```

Ensure that the functions are loaded on all processes.

```julia
@everywhere push!(LOAD_PATH, ".")
@everywhere using JuliaSet: julia_compute_domain
```

Compute the Julia set.

```julia
result, exec_procs = julia_compute_domain(-1.8 - 1.8im, 0.4 + 0.6im, 10_000, 4);
```

The distribution of the domains over the processes, remember that the first worker process has ID 2.

```julia
exec_procs
```

Visualize the result.

```julia
heatmap(result, aspect_ratio=:equal)
```
